```python
#loading the database
df1 = pd.read_csv("gtd.csv",encoding="ISO-8859-1",low_memory=False)
```

```python
#for a given attack, can we predict where it might occur?
#main columns would be country, city, iyear, imonth, iday, attacktype1, targtype1_txt, suicide, ransom, country_txt (class variable)
sampledDf = df1.sample(n=1000) #taking a sampled distribution of the data
#taking the columns we want
subSetOfSampled = sampledDf[['country','city','iyear','imonth','iday','attacktype1','targtype1_txt','suicide','ransom','country_txt']]
#this might be too complicated, might have to feed it into an ANN. but hell. It's worth a try
subSetOfSampled['country_txt']
```

```python
#write to arff file
#create a new file object
file = open('ForAGivenAttack.arff','w')
file.write("@RELATION ForAGivenAttack\n\n")
#for each column. write to a python .arff file

for columnName in sampledDf.columns:
    print(columnName)
    
file.write("@ATTRIBUTE {} REAL\n".format(subSetOfSampled.columns[0])) #country
file.write("@ATTRIBUTE {} STRING\n".format(subSetOfSampled.columns[1])) #city
file.write("@ATTRIBUTE {} REAL\n".format(subSetOfSampled.columns[2])) #iyear
file.write("@ATTRIBUTE {} REAL\n".format(subSetOfSampled.columns[3])) #imonth
file.write("@ATTRIBUTE {} REAL\n".format(subSetOfSampled.columns[4])) #iday
file.write("@ATTRIBUTE {} STRING\n".format(subSetOfSampled.columns[5])) #attacktype1
file.write("@ATTRIBUTE {} STRING\n".format(subSetOfSampled.columns[6])) #targtype1_txt
file.write("@ATTRIBUTE {} STRING\n".format(subSetOfSampled.columns[7])) #suicide
file.write("@ATTRIBUTE {} STRING\n".format(subSetOfSampled.columns[8])) #ransom

#go through all the countrys, put them into a list. 
#append it to a string and then add it below
#classVals = someFunct()

#make sure to increment row by one when done with adding a new value in
classVal = "@ATTRIBUTE " + subSetOfSampled.columns[9] + " {low, medium,high}"
file.write(classVal+"\n") #severity

file.write("@DATA\n")
#for each row in the sampled dataframe
for row in subSetOfSampled.iterrows():
    #for each value in the row, skip over the first one, which is the index
    writeAbleValue = ""
    index = row[0] #index is row 0
    values = row[1]#values are row 
    for value in reversed(row[1]): #had to reverse the order as the file was writing backwards
        print(value)
        writeAbleValue = (str(value) + ","+writeAbleValue)
    #print("\n\n")
    file.write(writeAbleValue+"\n")
#closing the file
file.close()
```
