```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import utils as ut
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)
```

```python
df1 = pd.read_csv("gtd.csv",encoding="ISO-8859-1",low_memory=False)
```

```python
df1.corr()
```

```python
#making a new casualties column
df1['casualties'] = df1['nkill'] + df1['nwound']
```

```python
#making a new severity column
casualties = df1['casualties'].fillna(0)
```

```python
df1['casualties'] = casualties
```

```python
#generating a new severity columns of zero data.
df1['severity'] = pd.Series(np.zeros(df1.shape[0]),index=df1.index)
```

```python
#generating severity given the number of casualties. simplified for checkpoint.
#future 
#low anything less than the mean.
#medium at the mean, up to one standard deviation
#high anything above 1 standard deviation

#gettting the mean of the casualty columns
meanOfDf = df1['casualties'].mean()
standardDeviation = df1['casualties'].std()
#select all the dataframe values where the casualty rate is less than the mean
lowSeverity = df1['casualties'][df1['casualties'] < meanOfDf]
#medium severity cases greater than the mean, but less than
#one standard deviation
mediumSeverity = df1['casualties'][(df1['casualties'] >= meanOfDf) & (df1['casualties'] < standardDeviation)]
#high severity instances are greater than one standard deviation
highSeverity = df1['casualties'][df1['casualties'] >= standardDeviation]

print(lowSeverity.mean())
print(mediumSeverity.mean())
print(highSeverity.mean())

df1['severity'][df1['casualties'] < meanOfDf] = 'low'
df1['severity'][(df1['casualties'] >= meanOfDf) & (df1['casualties'] < standardDeviation)] = 'medium'
df1['severity'][df1['casualties'] >= standardDeviation] = 'high'
df1['severity'].value_counts()
```
